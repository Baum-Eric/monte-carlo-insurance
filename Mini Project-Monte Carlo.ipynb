```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# number of simulations
n_sim = 10000

# average number of claims per year
lambda_claims = 50

# simulate number of claims
claims = np.random.poisson(lambda_claims, n_sim)
```

```python
# average claim size
mean_severity = 5000

# simulate severity (Gamma-like using exponential for simplicity)
severity = np.random.exponential(mean_severity, (n_sim, max(claims)))
```

```python
total_losses = []

for i in range(n_sim):
    n = claims[i]
    if n > 0:
        total_loss = np.sum(severity[i][:n])
    else:
        total_loss = 0
    total_losses.append(total_loss)

total_losses = np.array(total_losses)
```

```python
print("Average Loss:", np.mean(total_losses))
print("Std Dev:", np.std(total_losses))
print("Max Loss:", np.max(total_losses))
```

```python
plt.hist(total_losses, bins=50)
plt.title("Simulated Insurance Loss Distribution")
plt.xlabel("Total Loss")
plt.ylabel("Frequency")
plt.show()
```
